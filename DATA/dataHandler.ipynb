```python
import pandas as pd
```

```python
fonctionnaire = pd.read_csv('fonctionnaire.csv')
```

```python
fonctionnaire
```

```python
# First, convert all names to uppercase
fonctionnaire["Nom et prenom"] = fonctionnaire["Nom et prenom"].str.upper()

# Extract the last word as prenom (first name)
fonctionnaire["prenom"] = fonctionnaire["Nom et prenom"].str.split().str[-1]

# Remove the first name from the full name to get the last name
fonctionnaire["nom"] = fonctionnaire["Nom et prenom"].apply(lambda x: ' '.join(x.split()[:-1]))

# Show the results
fonctionnaire
```

```python
personnel = pd.read_excel('personnel.xls')
personnel
```

```python

def get_email(row):
    if pd.notna(row['Email Académique']) and row['Email Académique'].strip() != '':
        return row['Email Académique']
    elif pd.notna(row['Email Personnel']) and row['Email Personnel'].strip() != '':
        return row['Email Personnel']
    else:
        return 'invalid@usmba.ac.ma'

personnel['email'] = personnel.apply(get_email, axis=1)
```

```python
fonctionnaire['Nom et prenom'] = fonctionnaire['Nom et prenom'].str.upper()
personnel['Nom et prenom'] = personnel['Nom et prénom'].str.upper()
```

```python
# Merge fonctionnaire with personnel based on 'Nom et prenom'
fonctionnaire = fonctionnaire.merge(
    personnel[['Nom et prenom', 'email']], 
    on='Nom et prenom', 
    how='left'
)

# Fill missing emails with 'invalid@usmba.ac.ma'
fonctionnaire['email'] = fonctionnaire['email'].fillna('invalid@usmba.ac.ma')

# Show the results
print("Sample of merged data:")
fonctionnaire[['Nom et prenom', 'email']].head(10)
```

```python
fonctionnaire['telephone'] = 'N/A'
fonctionnaire['departement_id'] = fonctionnaire['Departement']
```

```python
# reorder columns
fonctionnaire = fonctionnaire[['nom', 'prenom','email','telephone','departement_id']]
fonctionnaire.to_csv('fonctionnaire2.csv',index=False)
```
